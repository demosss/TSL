```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
from statsforecast import StatsForecast
from statsforecast.models import Holt
from sklearn.metrics import mean_squared_error
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/demosss/TSL/main/data/SKAB/anomaly-free/anomaly-free.csv', sep=';')
```

```python
def create_sequences(data, column, seq_lenght):
    X, y = [], []
    for i in range(len(data) - seq_lenght):
        df_seq = data[['datetime', column]][i:i+seq_lenght]
        df_seq.columns = ['ds', 'y']
        df_seq['unique_id'] = column
        X.append(df_seq)
        y.append(data[column][i+seq_lenght])
    return X, np.array(y)    
```

### Accelerometer1RMS

```python
column = 'Accelerometer1RMS'
X, y_true = create_sequences(df, column, 30)
sf = StatsForecast(models = [Holt()], freq = 'S')
y_pred = []
for i in tqdm(range(len(X[:300]))):
    sf.fit(X[i])
    forecast_df = sf.forecast(h=1)
    y_pred.append(forecast_df['Holt'][0])

fig = plt.figure(figsize=(20, 8))
plt.plot(y_true[:300], label = "Истинные значения")
plt.plot(y_pred[:300], label = "Предсказанные значения")
plt.legend()
plt.show()

print("RMSE", column, mean_squared_error(y_true[:300], y_pred[:300]) ** 0.5)
```

```python
resudials = y_true[:300] - y_pred[:300]
sns.histplot(data=resudials, kde=True, bins=30)
plt.show()
```

# Полный расчет по всем временным рядам

```python
columns = ['Accelerometer1RMS', 'Accelerometer2RMS', 'Current', 'Pressure', 'Temperature', 'Thermocouple', 'Voltage', 'Volume Flow RateRMS']

for column in columns:
    print(column)

    X, y_true = create_sequences(df, column, 30)
    sf = StatsForecast(models = [Holt()], freq = 'S')
    y_pred = []
    for i in tqdm(range(len(X))):
        sf.fit(X[i])
        forecast_df = sf.forecast(h=1)
        y_pred.append(forecast_df['Holt'][0])

    print("RMSE", column, mean_squared_error(y_true, y_pred) ** 0.5)
    print()
```
